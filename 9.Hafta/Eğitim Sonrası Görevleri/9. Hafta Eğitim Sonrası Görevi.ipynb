# Problem 1

Kumanda isminde bir sınıf oluşturarak aşağıdaki işlemleri yapmasını sağlayan fonksiyonları sınıf içerisinde yazalım.
1. Tv Aç
2. Tv Kapat
3. Ses Ayarları
4. Kanal Ekle
5. Kanal Sayısını Öğrenme
6. Rastgele Kanala Geçme
7. Televizyon Bilgileri


Kumandayı kapatmak için q tuşuna basılmalı ve diğer durumlarda hangi işlemi yapmak istediği sorulmalıdır. Her işlemde gerekli fonksiyonunun sınıf içerisinden çağırılması gerekmektedir.

random ve time kütüphanesini araştırarak kullanabilirsiniz.

Kendi istediğiniz özellikleri eklemekte serbetsiniz.

```python
import random
import time

class kumanda:
    def __init__(self, tv_durum="kapalı", ses=10, kanal_listesi=["star"], mevcut_kanal="star"):
        self.tv_durum = tv_durum
        self.ses = ses
        self.kanal_listesi = kanal_listesi
        self.mevcut_kanal = mevcut_kanal

    def tv_ac(self):
        if self.tv_durum == "açık":
            print("televizyon zaten açık")
        else:
            self.tv_durum = "açık"
            print("televizyon açıldı")

    def tv_kapat(self):
        if self.tv_durum == "kapalı":
            print("televizyon zaten kapalı")
        else:
            self.tv_durum = "kapalı"
            print("televizyon kapatıldı")

    def ses_ayarla(self):
        yeni_ses = int(input("yeni ses seviyesini girin (0-100): "))
        if 0 <= yeni_ses <= 100:
            self.ses = yeni_ses
            print(f"ses seviyesi ayarlandı: {self.ses}")
        else:
            print("geçersiz ses seviyesi")

    def kanal_ekle(self, kanal):
        self.kanal_listesi.append(kanal)
        print(f"{kanal} kanalı eklendi")

    def kanal_sayisi(self):
        print(f"toplam kanal sayısı: {len(self.kanal_listesi)}")

    def rastgele_kanal_gec(self):
        self.mevcut_kanal = random.choice(self.kanal_listesi)
        print(f"yeni kanal: {self.mevcut_kanal}")

    def tv_bilgileri(self):
        print(f"tv durumu: {self.tv_durum}\nses seviyesi: {self.ses}\nmevcut kanal: {self.mevcut_kanal}\nkanallar: {self.kanal_listesi}")

def ana_menu():
    kumanda_obj = kumanda()

    while True:
        print("\n1. tv aç\n2. tv kapat\n3. ses ayarları\n4. kanal ekle\n5. kanal sayısını öğrenme\n6. rastgele kanala geçme\n7. televizyon bilgileri\nq. çıkış")
        secim = input("yapmak istediğiniz işlemi seçin: ").lower()

        if secim == "1":
            kumanda_obj.tv_ac()
        elif secim == "2":
            kumanda_obj.tv_kapat()
        elif secim == "3":
            kumanda_obj.ses_ayarla()
        elif secim == "4":
            kanal = input("eklemek istediğiniz kanal adı: ")
            kumanda_obj.kanal_ekle(kanal)
        elif secim == "5":
            kumanda_obj.kanal_sayisi()
        elif secim == "6":
            kumanda_obj.rastgele_kanal_gec()
        elif secim == "7":
            kumanda_obj.tv_bilgileri()
        elif secim == "q":
            print("çıkış yapılıyor...")
            time.sleep(1)
            break
        else:
            print("geçersiz işlem, tekrar deneyin")

ana_menu()
```
